# Parsing Raw Data

```python
# Load on every iteration of each chain
data = pd.DataFrame(columns=['logType','blockNumber','blockTimestamp','txnType','priceUsd','volumeUsd','amount0','amount1'])
```

```python
#change directory name to cronos, osmosis or canto to pull accirdingly
directory = 'cronos'

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    print(time.time())
    with open(f, 'r') as j:
        print(f)
        print(len(data))
        print(time.time())
        try:
            contents = json.loads(j.read())
        except UnicodeDecodeError:
            continue
        for i in range(0,len(contents['logs'])):
            #print(contents['logs'][i])
            try:
                if contents['logs'][i]['logType'] == 'swap':
                    data.loc[len(data)] = [contents['logs'][i]['logType'], 
                                             contents['logs'][i]['blockNumber'], 
                                             contents['logs'][i]['blockTimestamp'], 
                                             contents['logs'][i]['txnType'], 
                                             contents['logs'][i]['priceUsd'], 
                                             contents['logs'][i]['volumeUsd'],
                                             contents['logs'][i]['amount0'],
                                             contents['logs'][i]['amount1']
                                            ]
                else:
                    continue
            except KeyError:
                continue
            
            #contents_table = contents_table.reset_index(level=0)
            #contents_table.rename(columns={'index': 'delegator', 0: 'amount'}, inplace=True)
            #airdrop = airdrop.append(contents_table, ignore_index='True')
```

```python
#Evmos needs a different parsing logic becasue it has a different schema
directory = 'evmos'

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    print(time.time())
    with open(f, 'r') as j:
        print(f)
        print(len(data))
        print(time.time())
        try:
            contents = json.loads(j.read())
        except UnicodeDecodeError:
            continue
        for i in range(0,len(contents['tradingHistory'])):
            #print(contents['logs'][i])
            try:
                if (contents['tradingHistory'][i]['type'] == 'sell') or (contents['tradingHistory'][i]['type'] == 'buy'):
                    data.loc[len(data)] = [contents['tradingHistory'][i]['type'], 
                                             contents['tradingHistory'][i]['blockNumber'], 
                                             contents['tradingHistory'][i]['blockTimestamp'], 
                                             contents['tradingHistory'][i]['type'], 
                                             contents['tradingHistory'][i]['priceUsd'], 
                                             contents['tradingHistory'][i]['volumeUsd'],
                                             contents['tradingHistory'][i]['amount0'],
                                             contents['tradingHistory'][i]['amount1']
                                            ]
                else:
                    continue
            except KeyError:
                continue
            
            #contents_table = contents_table.reset_index(level=0)
            #contents_table.rename(columns={'index': 'delegator', 0: 'amount'}, inplace=True)
            #airdrop = airdrop.append(contents_table, ignore_index='True')
```

```python
def time_of_opp(tolerance, dataset):
    counter = 0
    opps = []
    for item in dataset:
        if item >= tolerance:
            counter += 1
        elif item <= tolerance and counter != 0:
            opps.append(counter)
            counter = 0
        else:
            counter = 0
    print('opps:' +str(len(opps)))
    print('average opp duration:' +str(sum(opps)/len(opps)))
    return opps
```
